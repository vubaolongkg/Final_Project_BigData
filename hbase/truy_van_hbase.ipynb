```python
import happybase

connection = happybase.Connection('localhost', port=9090)
table = connection.table('real_estate')

for key, data in table.scan():
    try:
        gia = float(data.get(b'info:Gia', b'0'))
        phong_ngu = int(data.get(b'info:PhongNgu', b'0'))

        if gia < 5 and phong_ngu > 3:
            print(f"{key.decode()}: Gia={gia}, PhongNgu={phong_ngu}")
    except ValueError:
        continue
```

```python
import happybase

connection = happybase.Connection('localhost', port=9090)
table = connection.table('real_estate')

for key, data in table.scan():
    try:
        tinh_thanh = data.get(b'info:TinhThanh', b'').decode()
        wc = int(data.get(b'info:WC', b'0'))

        if tinh_thanh == 'Hà Nội' and 2 <= wc <= 4:
            print(f"{key.decode()}: WC={wc}, Tỉnh/Thành={tinh_thanh}")
    except ValueError:
        continue
```

```python
import happybase

connection = happybase.Connection('localhost', port=9090)
table = connection.table('real_estate')

# Danh sách chứa các bản ghi (key, gia, data)
records = []

for key, data in table.scan():
    try:
        gia = float(data.get(b'info:Gia', b'0'))
        records.append((key, gia, data))
    except ValueError:
        continue

# Sắp xếp theo giá giảm dần
records.sort(key=lambda x: x[1], reverse=True)

# In ra top 5
print("Top 5 bất động sản có giá cao nhất:")
for i, (key, gia, data) in enumerate(records[:5], start=1):
    tieu_de = data.get(b'info:TieuDe', b'').decode()
    print(f"{i}. {tieu_de} - Giá: {gia} tỷ (RowKey: {key.decode()})")
```

```python
import happybase

connection = happybase.Connection('localhost', port=9090)
table = connection.table('real_estate')

# Danh sách chứa các bản ghi (key, gia, data)
records = []

for key, data in table.scan():
    try:
        gia = float(data.get(b'info:Gia', b'0'))
        records.append((key, gia, data))
    except ValueError:
        continue

# Sắp xếp theo giá tăng dần
records.sort(key=lambda x: x[1])

# In ra top 5
print("Top 5 bất động sản có giá rẻ nhất:")
for i, (key, gia, data) in enumerate(records[:5], start=1):
    tieu_de = data.get(b'info:TieuDe', b'').decode()
    print(f"{i}. {tieu_de} - Giá: {gia} tỷ (RowKey: {key.decode()})")
```

```python
import happybase

connection = happybase.Connection('localhost', port=9090)
table = connection.table('real_estate')

records = []

for key, data in table.scan():
    try:
        gia_m2 = float(data.get(b'info:GiaTrenM2', b'0'))
        records.append((key, gia_m2, data))
    except ValueError:
        continue

# Sắp xếp giảm dần theo giá trên m²
records.sort(key=lambda x: x[1], reverse=True)

# In ra top 5
print("Top 5 bất động sản có giá trên mỗi mét vuông cao nhất:")
for i, (key, gia_m2, data) in enumerate(records[:5], start=1):
    tieu_de = data.get(b'info:TieuDe', b'').decode()
    dien_tich = data.get(b'info:DienTich', b'').decode()
    print(f"{i}. {tieu_de} - {gia_m2} triệu/m² - Diện tích: {dien_tich}m² (RowKey: {key.decode()})")
```
